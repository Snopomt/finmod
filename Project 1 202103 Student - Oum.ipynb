## Step 5a Construct a function envelope(c)  to calculate risk, return and proportion of stocks in envelope portfolio. 
*As mention in the workshop, construct an function called 'envelope(c)' to calculate the envelope portfolio from a given constant $c$. You should use the matrix R and S to build this function, and return risk, return, and proportion of each stocks in the portfolio.* 

*To use the inverse function you should import this function from 'from numpy.linalg import inv'*


```python
from numpy.linalg import inv

def envelope(c):
    E = np.dot(inv(S),R-c)/sum(np.dot(inv(S),R-c))
    rE = np.dot(E,R)
    sE = np.sqrt(np.dot(np.dot(E,S),E))
    return np.array([sE,rE,E])
```

## Step 6 Construct an envelope
*After having two envelope portfolios, you can construct portfolio $Z$, from both envelope portfolios and an $\alpha$ term. Construct an envelope.*

```python
rx = rX
ry = rY
sigmax = sX
sigmay = sY

X = envelope(cX)[2]
Y = envelope(cY)[2]
sigma_xy = np.dot(np.dot(X,S),Y)

Z = pd.DataFrame([{'alpha': 0, \
                   'sigmap': 0, \
                   'rp': 0}])

RangeAlpha = np.arange(-1.0,1.1,0.1)

for alpha in RangeAlpha:
    Z = Z.append({'alpha': alpha, \
                  'sigmap': (alpha**2*sigmax**2 + \
                            (1-alpha)**2*sigmay**2 + \
                            2*alpha*(1-alpha)*sigma_xy)**0.5, \
                  'rp': alpha*rx + (1-alpha)*ry}, \
                 ignore_index = True)
Z = Z.drop(labels=0, axis=0)
```

## Step 8 Draw the efficient frontier
*Instead of using portfolio X and Y to draw an envelope, here you use portfolio X and GMVP to constuct the efficient frontier. Repeat drawing the Z portfolio again.*

```python
calculate input for constructing portfolio Z, return and risk of portfolio X and Y
rx = rX
ry = rGMVP
sigmax = sX
sigmay = sGMVP

 calculate covariance of portfolio X and Y
X = envelope(cX)[2]
Y = envelope(cY)[2]
sigma_xy = np.dot(np.dot(X,S),GMVP)

 create DataFrame of portfolio Z, including alpha term, risk, and return of portfolio Zi
Z = pd.DataFrame({['alpha': 0, 'sigma':0, 'rp':0]})

 create a range of alpha term. Range of alpha should create list of portfolio Z that cover all risky assets. 
RangeAlpha = np.arange(0,1.1,0.1)

 create Z portfolio respective the range of alpha, as well as droping the first row data.
for alpha in RangeAlpha:
    Z = Z.append({'alpha': alpha, \
                  'sigmap': (alpha**2*sigmax**2 + (1-alpha)**2*sigmay**2 + 2*alpha*(1-alpha)*sigma_xy)**0.5, \
                  'rp':alpha*rx + (1-alpha)*ry}, \
                 ignore_index = True)
Z = Z.drop(labels=0, axis=0)
```
